```python
data = list()
files = ['split_ranking_0.csv','split_ranking_1.csv','split_ranking_2.csv','split_ranking_3.csv','split_ranking_4.csv']
for file in files:
  file = '../stacked_data/' + file
  with open(file, 'r', newline='') as f:
    reader = csv.reader(f, delimiter=',')
    headers = next(reader, None)
    print(headers)
    sublist = list()
    cur_group = 0
    list_idx = [0, 4, 5]
    int_idx = [1,2,3,6]
    for row in reader:
      for i in list_idx:
        temp = list()
        for a in row[i].replace('[','').replace(']','').split(','):
          temp.append(float(a))
        row[i] = temp
      for i in int_idx:
        if row[i] == '':
          row[i] = -1
        row[i] = np.float(row[i])
      if row[-1] != cur_group:
        data.append(sublist)
        sublist = list()
        cur_group += 1
      sublist.append(row)
```

```python
len(data)
```

```python
from copy import deepcopy
test = deepcopy(data)
```

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
#bleu feature?
positives = list()
negatives = list()
for beam in test:
  target = beam[0][-2]
  for ex in beam:
    ex.append(sentence_bleu([target], ex[0]))
  beam.sort(reverse=True, key=lambda x: x[-1])
  beam[0][-1] = 1
  for sub_b in beam[1:]:
    sub_b[-1] = 0
  for _ in range(len(beam)-1):
    positives.append(beam[0])
  for sub_b in beam[1:]:
    negatives.append(sub_b)
  print(len(positives), len(negatives))
```

```python
# create dataset of positional_scores + score
largest = max([len(a[4]) for a in positives])
positive_np = np.zeros((len(positives), largest+1))
for i, p in enumerate(positives):
  additive = deepcopy(p[4])
  additive.append(p[1])
  positive_np[i,:len(additive)] += additive
largest = max([len(a[4]) for a in negatives])
negative_np = np.zeros((len(negatives), largest+1))
for i, n in enumerate(negatives):
  additive = deepcopy(n[4])
  additive.append(n[1])
  negative_np[i,:len(additive)] += additive
```
